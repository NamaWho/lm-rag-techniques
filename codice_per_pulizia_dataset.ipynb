## Come inserire questa cella nel codice main?
Inserire questa cella tra la cella in cui carichiamo il dataset con 
```python
dataset = ir_datasets.load("cord19/trec-covid")
```
e la cella che inizia per
```python
# Prepare documents and queries
print("Preparing documents and queries...")

# put all documents and queries in a list of dictionaries
all_docs = []
for doc in dataset.docs_iter():
    ...
```
questa cella andrà modificata in quanto non si dovrà iterare più su `dataset.docs_iter()` ma sulla list `all_docs` (che viene ora inizializzata in questa nuova cella, bisogna quindi togliere `all_docs = []`).

PS: dovrebbe funzionare, ovviamente andrà fatto qualche minimo aggiustamento di tipi.

````python
def remove_duplicates_by_abstract(dataset):
    seen_abstracts = set()
    duplicate_count = 0
    unique_documents = []

    for document in dataset:
        abstract = document.abstract
        if abstract not in seen_abstracts:
            seen_abstracts.add(abstract)
            unique_documents.append(document)
        else:
            duplicate_count += 1

    return unique_documents, duplicate_count

def remove_duplicated_by_title(dataset):
    seen_titles = set()
    duplicate_count = 0
    unique_documents = []

    for document in dataset:
        title = document.title
        if title not in seen_titles:
            seen_titles.add(title)
            unique_documents.append(document)
        else:
            duplicate_count += 1

    return unique_documents, duplicate_count


# divido in due liste i documenti con abstract e senza abstract
docs_with_abstract = []
docs_without_abstract = []

for doc in dataset.docs_iter():
    if doc.abstract:
        docs_with_abstract.append(doc)
    else:
        docs_without_abstract.append(doc)

print(f"Number of documents with abstract: {len(docs_with_abstract)}")
print(f"Number of documents without abstract: {len(docs_without_abstract)}")
print(f"Total number of documents: {len(docs_with_abstract) + len(docs_without_abstract)}", "\n")

unique_documents_by_abstract, duplicate_abstract_count = remove_duplicates_by_abstract(docs_with_abstract)
print(f"Number of unique documents by abstract: {len(unique_documents_by_abstract)}")
print(f"Number of duplicate documents by abstract: {duplicate_abstract_count}")
assert len(unique_documents_by_abstract) + duplicate_abstract_count == len(docs_with_abstract), "Errore"

unique_documents_by_title, duplicate_title_count = remove_duplicated_by_title(docs_without_abstract)
print(f"Number of unique documents by title: {len(unique_documents_by_title)}")
print(f"Number of duplicate documents by title: {duplicate_title_count}")
assert len(unique_documents_by_title) + duplicate_title_count == len(docs_without_abstract), "Errore"

assert len(unique_documents_by_abstract) + len(unique_documents_by_title) + duplicate_abstract_count + duplicate_title_count == dataset.docs_count(), "Errore"

all_docs = unique_documents_by_abstract + unique_documents_by_title
````
